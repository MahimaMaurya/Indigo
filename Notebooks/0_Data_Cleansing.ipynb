# Data Cleaning

Spelling errors - `Accuracy`


Irrrelavant HTML tags - `Validity`

Emojis - `Validity`

Presence of \n  - `Consistency`



> Data Cleaning Order
1. Quality -> Completeness
2. Tidiness
3. Quality -> Validity
4. Quality -> Accuracy
5. Quality -> Inconsistency



```python
train_df1 = train_df.copy()
```

```python
import re
```

```python
def remove_html_tags(text):
    return re.sub(r'<.*?>', '', text)

# Apply HTML tag removal
train_df1['question'] = train_df1['question'].apply(remove_html_tags)
train_df1['answer'] = train_df1['answer'].apply(remove_html_tags)
```

```python
# Compile the emoji pattern
EMOJI_PATTERN = re.compile(
    "(["
    "\U0001F1E0-\U0001F1FF"  # flags (iOS)
    "\U0001F300-\U0001F5FF"  # symbols & pictographs
    "\U0001F600-\U0001F64F"  # emoticons
    "\U0001F680-\U0001F6FF"  # transport & map symbols
    "\U0001F700-\U0001F77F"  # alchemical symbols
    "\U0001F780-\U0001F7FF"  # Geometric Shapes Extended
    "\U0001F800-\U0001F8FF"  # Supplemental Arrows-C
    "\U0001F900-\U0001F9FF"  # Supplemental Symbols and Pictographs
    "\U0001FA00-\U0001FA6F"  # Chess Symbols
    "\U0001FA70-\U0001FAFF"  # Symbols and Pictographs Extended-A
    "\U00002702-\U000027B0"  # Dingbats
    "])"
)

def remove_emojis(text):
    return EMOJI_PATTERN.sub(r'', text)

    # Apply emoji removal to the 'question' and 'answer' columns
train_df1['question'] = train_df1['question'].apply(remove_emojis)
train_df1['answer'] = train_df1['answer'].apply(remove_emojis)
```

```python
def remove_newlines(text):
    return text.replace('\n', ' ')

# Apply the function to your dataset
train_df1['question'] = train_df1['question'].apply(remove_newlines)
train_df1['answer'] = train_df1['answer'].apply(remove_newlines)
```

```python
# Save the cleaned DataFrame to a CSV file
train_df1.to_csv('/content/cleaned_train_df.csv', index=False)

# Load the cleaned DataFrame from the CSV file
cleaned_df = pd.read_csv('cleaned_train_df.csv')

# Display the first few rows
print(cleaned_df.head())
```
